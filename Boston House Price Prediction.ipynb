## Exploratory Data Analysis

In case of a regression problem statement, with the features as numerical, the first code to run is correlation. It is very important as we need to find how the independent features and the dependent features are correlated. Model performance would be high if they are highly positively or highly negatively correlated.

```python
plt.figure(figsize=(10,6))
sns.heatmap(dataset.corr(), annot=True)
plt.show()
```

Here we check the correlation between
1. among the independent features
2. independent feature and the dependent feature

a.  If there is high correlation between independent features, then one can remove one of the independent features. This is called a s multicollinearity. Multicollinearity: If two independent features are highly positively or highly negatively correlated. Around 90 to 95%. We can drop one of the independent features. <br>
b.  Check the correlation of independent variables with dependent variable

```python
sns.pairplot(dataset)
```

## Analyse the correlated features

We have to analyse with respect to dependent and independent features.

```python
for feature in dataset.columns:
    if feature!='Price':
        sns.scatterplot(x=feature, y='Price', data=dataset)
        plt.title(feature)
        plt.show()
    
```

Observation:
1. We see that as Crime rate increases the house price decreases.
2. With increase in RM(Average number of rooms per acre) the price increases.
3. With age, the price decreases.
4. With increase in DIS (weighted distances to five Boston employment centres), the price increases.
5. With LSTAT( % lower status of the population) the price decreases. As the satatus of people becomes more low, the price decreases.


```python
#Create regression plot for the data with above columns
feature_select=['CRIM', 'RM', 'AGE', 'DIS', 'LSTAT']
for feature in feature_select:
    sns.regplot(x=feature, y='Price', data=dataset)
    plt.show()
```

To create a better regression model, there should be some correlation present between the feature and the target or dependent feature. Positive or negative linearity.

Here we are going to take up all the independent features.

```python
# Independent and dependent variable
# All the independent features are stored in X. We have to take all the features except the last column the price.
# in y we take only the Price column
X=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

```python
X.head()
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.3, random_state=23)
```

We would not be touching the test data. For prediction and how the model is performing we would use the test data.

#### Standardise the dataset

In linear regression we use gradient descent. Our aim is to reach the global minima. Every feature is calculated with respect to different units. In order for the converging of the gradient descent to take faster, we have to standardise/normalise all the datapoints to the same scale. Standard Scaler/ Ztransform into somewhat normal distribution and gives units in terms of how much the standard deviation away the particular datapoint is.

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

We have to transform the X_train dataset completely to the same scale. But for X_test, we would not write fit_transform so that the model does not know much information about the test dataset. The test data would not be fitted to a model. We do not have to do it for the output feature. the y_train and y_test.

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
import pickle
pickle.dump(scaler, open('scaling.pkl','wb'))
```

## Model Training

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
```

```python
regressor.fit(X_train,y_train)
```

```python
#print the coefficients and the intercept
print(regressor.coef_)
print('The intercept is', regressor.intercept_)
```

```python
#On which parameters the model has been trained
regressor.get_params()
```

Now we would make prediction with respect to test data in order to see how the model performs. Later we would make predictions on new data. The new data would have to be transformed with Standard Scaler

```python
#Prediction with the test data.
reg_predictions=regressor.predict(X_test)
```

```python
reg_predictions
```

To check the kind of predictions whether they are good or bad, we would be plotting some graphs to understand how well the model performed.

```python
#Plotting a scatter plot for the prediction
plt.scatter(y_test, reg_predictions)
plt.xlabel('y_test')
plt.ylabel('Predictions')
```

In the scatterplot, the plotting is basically linear. It means that the model has performed well. With the increase in y_test values, the predictions are also increasing. So a positive sign.

```python
#Plotting the residuals or the errors
residuals=y_test-reg_predictions
sns.distplot(residuals, hist=False, kde=True)
```

The plot of the residuals looks like a normal distribution. There are some outliers to both right and left side of the curve but the errors are not so high. From the scatterplot also we see that some points are far from the cluster both on the high end(Above the cluster) as well as low end(below the cluster).

When we are trying to solve the problem our aim should be that we should try to get this as a normal distribution. Here most of the points have an error between -10 to +10.

```python
#Scatterplot with respect to residual and prediction
plt.scatter(reg_predictions, residuals)
plt.xlabel('Predictions')
plt.ylabel('Residuals')
```

We see that there is a uniform distribution of residuals vs predictions plot. No specific trend. It is good for the model. If there was some trend like with increasing values of the prediction, the residuals are also increasing then it would mean that the model was not performing well for high values  and it was not predicting well.

Most predictions have residuals near zero.

### Performance Metrics

```python
#Finding mean squared error or cost function. Mean absolute error is used in Lasso. These should be lesser
from sklearn.metrics import mean_squared_error, mean_absolute_error
mse=mean_squared_error(y_test, reg_predictions)
abmse=mean_absolute_error(y_test, reg_predictions)
print('Mean absolute error:', abmse)
print('Mean squared error:', mse)
print('Root mean squared error', np.sqrt(mse))
```

##### R square and adjusted Rsquare.

```python
# The closer the values are to 1, the better the model is performing
from sklearn.metrics import r2_score
score=r2_score(y_test, reg_predictions)
print(score)
```

```python
#Adjusted R2. It would be less than R2
adjusted_r2score=1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
print(adjusted_r2score)
```
